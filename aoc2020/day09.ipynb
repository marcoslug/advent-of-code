### Input

```python
with open("data/09_input.txt") as f:
    l = f.read().splitlines()
```

```python
l = [int(n) for n in l]
```

### Part 1

```python
for i in range(25, len(l)+1):
    n = l[i]
    print(f"Considering {i+1}th number, which is <{n}>.")
    found = False
    j = i - 25
    while (not found) and (j < i - 1):
        k = j + 1
        a = l[j]
        while (not found) and (k < i):
            b = l[k]
            found = (n == (a + b))
            k += 1
        j += 1
    if found:
        print(f"Ok, I found <{a}> ({j}th number) and <{b}> ({k}th number) that sum up to <{n}>.")
    else:
        print(f"No addends found for this number!")
        break        
```

### Part 2

```python
pos = 500
n_target = l[pos]
```

```python
found = False
for length in range(2, pos+1):
    print(f"Considering sequences of {length} numbers...")
    i_start = 0
    while (not found) and ((i_start + length) <= pos):
        l_red = l[i_start:(i_start+length)]
        found = (sum(l_red) == n_target)
        i_start += 1
    if found:
        print("A combination has been found!")
        break
if found:
    print("Here are the numbers of the combination: " + ", ".join([str(n_red) for  n_red in l_red]) + ".")
else:
    print("No combination found.")
```
