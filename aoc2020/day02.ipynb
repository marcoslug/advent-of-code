```python
import numpy as np
```

### Input

```python
with open("data/02_input.txt") as f:
    data = f.read().splitlines()
```

```python
data
```

### Part 1: check valid passwords

```python
cnt_valid = 0
for l in data:
    parts = l.split(": ")
    pwd = parts[1]
    left = parts[0].split(" ")
    letter = left[1]
    nums = left[0].split("-")
    pwd_array = np.array([c for c in pwd])
    letter_count = sum(pwd_array==letter)
    cnt_valid += int(letter_count in range(int(nums[0]), int(nums[1])+1))
```

```python
cnt_valid
```

### Part 2: new policy

```python
cnt_valid = 0
for l in data:
    parts = l.split(": ")
    pwd = parts[1]
    left = parts[0].split(" ")
    letter = left[1]
    nums = left[0].split("-")
    pwd_array = np.array([c for c in pwd])
    letter_count = sum(pwd_array==letter)
    cnt_same = int(pwd_array[int(nums[0])-1] == letter) + int(pwd_array[int(nums[1])-1] == letter)
    cnt_valid += int(cnt_same == 1)
```
